# Flight Price Prediction (EDA + Feature Engineering)

```python
#importig Basics libraries
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline 
# %matplotlib inline : This is used for if you want to show a graph without using plt.show()
```

```python
#Imort Train datset
Train_dataset=pd.read_excel("Data_Train.xlsx")
Train_dataset.head(3)
```

```python
# Import Test Datset
Test_Datset=pd.read_excel("Test_set.xlsx")
Test_Datset.head(3)
```

```python
#Combinining Both table
Final_dataset=Train_dataset.merge(Test_Datset,"left")
Final_dataset.head(4)
```

```python
#Feature engineering Process
Final_dataset['Date']=Final_dataset['Date_of_Journey'].str.split('/').str[0]
Final_dataset['Month']=Final_dataset['Date_of_Journey'].str.split('/').str[1]
Final_dataset['Year']=Final_dataset['Date_of_Journey'].str.split('/').str[2]
Final_dataset['Arrival_Time']=Final_dataset['Arrival_Time'].str.split(' ').str[0]
Final_dataset['Arrival_Hour']=Final_dataset['Arrival_Time'].str.split(':').str[0]
Final_dataset['Arrival_Mint']=Final_dataset['Arrival_Time'].str.split(':').str[1]
Final_dataset['Dep_Hour']=Final_dataset['Dep_Time'].str.split(':').str[0]
Final_dataset['Dept_Mint']=Final_dataset['Dep_Time'].str.split(':').str[1]
Final_dataset['Duration_Hour']=Final_dataset['Duration'].str.split(' ').str[0].str.split('h').str[0]

Final_dataset.head()
```

```python
# Convert the data type from object to integer
Final_dataset['Date']=Final_dataset['Date'].astype(int)
Final_dataset['Month']=Final_dataset['Month'].astype(int)
Final_dataset['Year']=Final_dataset['Year'].astype(int)
Final_dataset['Arrival_Hour']=Final_dataset['Arrival_Hour'].astype(int)
Final_dataset['Dep_Hour']=Final_dataset['Dep_Hour'].astype(int)
Final_dataset['Dept_Mint']=Final_dataset['Dept_Mint'].astype(int)
Final_dataset['Arrival_Mint']=Final_dataset['Arrival_Mint'].astype(int)
Final_dataset.columns
```

```python
#Remove the columne of which is not use ful for preciction 
Final_dataset.drop("Date_of_Journey",axis=1,inplace=True)
Final_dataset.drop("Arrival_Time",axis=1,inplace=True)
Final_dataset.drop('Dep_Time',axis=1,inplace=True)

Final_dataset.columns
```

```python
# This is used for mapping the the indiviual column for seperating the string
Final_dataset['Total_Stops'].unique()
Final_dataset['Total_Stops']=Final_dataset['Total_Stops'].map({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3,'4 stops':4,'nan':1})
Final_dataset.drop('Route',axis=1,inplace=True)
Final_dataset.head(5)
```

```python
# Delete those column who is in 
Final_dataset[Final_dataset['Duration_Hour']=='5m']
Final_dataset.drop(6480,inplace=True,axis=0)
Final_dataset.drop('Duration',axis=1,inplace=True)
Final_dataset['Duration_Hour']=Final_dataset['Duration_Hour'].astype(int)
Final_dataset.dtypes
# Data is ready for prediction 
```

```python
Final_dataset['Airline'].unique()
```

```python
# We will use train and test data then important is traform a techinque
from sklearn.preprocessing import LabelEncoder 

# fit()= Learns the parameters from the data (e.g., mean and standard deviation for scaling).
# transform()= Applies the learned parameters to transform the data.

Label_encoder=LabelEncoder()
Final_dataset['Airline']=Label_encoder.fit_transform(Final_dataset['Airline'])
Final_dataset['Source']=Label_encoder.fit_transform(Final_dataset['Source'])
Final_dataset['Destination']=Label_encoder.fit_transform(Final_dataset['Destination'])
Final_dataset['Additional_Info']=Label_encoder.fit_transform(Final_dataset['Additional_Info'])
```
